```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import geopandas as gpd
import matplotlib.animation as animation
import pandas as pd
import shapely
from shapely.geometry import Polygon
from shapely.geometry import Point
from shapely.wkt import loads
import geojson
import requests
from scipy.ndimage import convolve
from matplotlib.animation import FuncAnimation, FFMpegWriter
```

```python
geojson_file = "smokeSimFinal.geojson"
gdf = gpd.read_file(geojson_file)
geojson_file2 = 'Boundaries-Chicago.geojson' 
gdf2 = gpd.read_file(geojson_file2)
geojson_file3 = 'Communities-Chicago.geojson'
gdf3 = gpd.read_file(geojson_file3)
grid_size = (200, 200) 

if 'smoke_density' in gdf.columns:
    gdf['smoke_density'] = gdf['smoke_density'].apply(lambda x: 0 if x < 0.00008 else x)
```

```python
fig, ax = plt.subplots(figsize=(50, 50))
ax.set_xlim(0, grid_size[1])
ax.set_ylim(0, grid_size[0])

ax.clear()


gdf.plot(ax=ax, edgecolor='black', linewidth=0.01, facecolor='white')
mask = gdf['smoke_density'] >= 0.00008
gdf[mask].plot(ax=ax, column='smoke_density', cmap='jet', legend=True)
gdf[~mask].plot(ax=ax, edgecolor='black', linewidth=0.1, facecolor='white')
gdf2.plot(ax=ax, edgecolor='red', linewidth=0.5, facecolor='none')
gdf3.plot(ax=ax, edgecolor='green', linewidth=0.5, facecolor='none')
```

```python
print(gdf.crs)
print(gdf3.crs)
```

```python
fig, ax = plt.subplots(figsize=(50, 50))
ax.set_xlim(0, grid_size[1])
ax.set_ylim(0, grid_size[0])

ax.clear()
gdf1_with_community.plot(ax=ax, column='color_index', cmap='jet', legend=True)
```

```python

gdf1_with_community['community'] = gdf1_with_community['community'].fillna('None')

y
community_summary = gdf1_with_community.groupby('community').agg(
    total_smoke_density=('smoke_density', 'sum'),
    geometry=('geometry', 'first')
).reset_index()


gdf_community_summary = gpd.GeoDataFrame(community_summary, geometry='geometry', crs=gdf1_with_community.crs)

gdf_community_summary.to_file("community_summary.geojson", driver="GeoJSON")

print(gdf_community_summary)
```

```python

gdf_combined = gdf3.merge(
    gdf_community_summary[['community', 'total_smoke_density']], 
    on='community', 
    how='left'
)

gdf_combined['total_smoke_density'] = gdf_combined['total_smoke_density'].fillna(0)

gdf_combined = gpd.GeoDataFrame(gdf_combined, geometry='geometry', crs=gdf3.crs)

print(gdf_combined)
```

```python
fig, ax = plt.subplots(figsize=(50, 50))
ax.set_xlim(0, grid_size[1])
ax.set_ylim(0, grid_size[0])

ax.clear()
gdf_combined.plot(
    ax=ax, 
    column='total_smoke_density', 
    cmap='jet', 
    legend=True # Adjust font size here
)
```

```python

gdf_smoke = gpd.read_file("smokeSimFinal.geojson") 
gdf_schools = gpd.read_file("elementary_schools.geojson")  


gdf_smoke = gdf_smoke.to_crs(gdf_schools.crs)


gdf_joined = gpd.sjoin(gdf_schools, gdf_smoke, how="left", predicate="intersects")


gdf_schools['smoke_density'] = gdf_joined['smoke_density']

print(gdf_schools.head())
```

```python

fig, ax = plt.subplots(figsize=(12, 10))
gdf_schools.plot(
    column='smoke_density', 
    cmap='RdYlBu',           
    legend=True,             
    markersize=50,           
    ax=ax,
    alpha=0.8,              
)


plt.title("Elementary Schools Colored by Smoke Density (Gradient Colormap)", fontsize=16)
plt.xlabel("Longitude")
plt.ylabel("Latitude")


plt.show()
```

```python

gdf_schools_sorted = gdf_schools.sort_values(by='smoke_density', ascending=False).head(20)
non_zero_schools = gdf_schools[gdf_schools['smoke_density'] > 10]

plt.figure(figsize=(30, 10))
plt.bar(
    non_zero_schools['Short_Name'], 
    non_zero_schools['smoke_density'], 
    color='red', alpha=0.8
)


plt.title("Smoke Density at Elementary School Locations", fontsize=16)
plt.xlabel("Elementary Schools", fontsize=12)
plt.ylabel("Smoke Density", fontsize=12)
plt.xticks(rotation=90) 
plt.tight_layout()


plt.show()
```
